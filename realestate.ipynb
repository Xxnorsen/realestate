```python
import pandas as pd
import os

# Use the local path for VS Code
csv_path = 'data.csv'

# Check if file exists in current directory
if os.path.exists(csv_path):
    print(f"‚úÖ Found file at: {csv_path}")
    df = pd.read_csv(csv_path)
    display(df.head(20))
else:
    print(f"‚ùå File not found at: {csv_path}")
    print(f"Current directory: {os.getcwd()}")
    print("Please make sure data.csv is in the same directory as this notebook")
    
    # List files in current directory to help debug
    print("\nFiles in current directory:")
    for file in os.listdir('.'):
        if file.endswith('.csv'):
            print(f"  - {file}")
```

```python
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (12, 6)

print("=" * 60)
print("STEP 2: DATA EXPLORATION")
print("=" * 60)


# 1. Basic Information
print("\nüìä DATASET SHAPE:")
print(f"   Rows: {df.shape[0]:,}")
print(f"   Columns: {df.shape[1]}")

# 2. Column Information
print("\nüìã COLUMN DETAILS:")
print(df.info())

# 3. Data Types
print("\nüî¢ DATA TYPES:")
print(df.dtypes)

# 4. Statistical Summary
print("\nüìà STATISTICAL SUMMARY:")
print(df.describe())

# 5. Missing Values Check
print("\n‚ùì MISSING VALUES:")
missing = df.isnull().sum()
missing_pct = (df.isnull().sum() / len(df)) * 100
missing_df = pd.DataFrame({
    'Column': missing.index,
    'Missing_Count': missing.values,
    'Percentage': missing_pct.values
})
missing_df = missing_df[missing_df['Missing_Count'] > 0].sort_values('Missing_Count', ascending=False)

if len(missing_df) > 0:
    print(missing_df.to_string(index=False))
else:
    print("‚úÖ No missing values found!")
```
